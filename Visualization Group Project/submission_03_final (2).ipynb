#### Visualization

#### Description

This visualization implies that people with the lowest evaluation score are most likely to get promoted.

Obviously this is not true. This plot is misleading because there is no information on either when the promotion happened or when the evaluation took place. In this case the evaluation score describes the perfomance over the last year while the timespan of the promotion is much larger, namely five years, which indicates that it is likely that the evalution took place after the promotion. Because the evaluation of people in management cannot be compared to people in other departments the plot is misleading. Additionally, we cut the y-axis before 1.5 to make the differences appear even larger.
